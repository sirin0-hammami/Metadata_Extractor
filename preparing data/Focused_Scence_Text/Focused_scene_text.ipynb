Plot bounding box on a sample

```python
import matplotlib.pyplot as plt 
from matplotlib.patches import Rectangle
import cv2
```

```python
def draw_bounding_box(img_path, x, y, w, h):
    
    # Create figure and axes
    fig,ax = plt.subplots(1)
    img=cv2.imread(img_path)
    # Display the image
    ax.imshow(img)

    rect = Rectangle((x,y),w, h,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)

    plt.show()
```

```python
draw_bounding_box("/home/sirine/fiftyone/data/Focused Scene Text/images/train/196.jpg", 1200 ,755, 2561-1200, 895-755 )
```
