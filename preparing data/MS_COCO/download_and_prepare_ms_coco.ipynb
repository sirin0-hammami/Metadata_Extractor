Exctacting classes and corerspondant ids

```python
#extracting classes and ids in a dict 
dict = {}

for cat in data["categories"]:
    dict[cat["id"]] = cat["name"]
```

```python
keys = list(dict.keys())
for i in keys:
    print(dict[i])
```

```python
#writing the classes and ids in a text file 
f = open("/home/sirine/fiftyone/preparing data/convert_mscoco/classes.txt", "w")
for i in keys:
    f.write(f"{i} {dict[i]} \n")
f.close()
```
